## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests as req
import json

pop = req.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
none_pop = req.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')

response = pop.text

print(response)
```

```python
import json

pop_data = json.loads(pop.text)
print ('這個 API 一次會回傳',len(pop_data),'筆資料')

json.loads(pop.text)
    
```

```python

nonepop_data = json.loads(none_pop.text)

json.loads(none_pop.text)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
pop_message = 0
pop_like = 0
for i in pop_data:
    print('標題    : ',i['title'])
    print('貼文時間 : ',i['createdAt'])
    print('留言人數 : ',i['commentCount'])
    print('按讚人數 : ',i['likeCount'])
    print('=============')
    pop_message = pop_message + i['commentCount']
    pop_like = pop_like + i['likeCount']

print('熱門留言總數',pop_message)
print('熱門按讚總數',pop_like)
```

```python
nonepop_message = 0
nonepop_like = 0
for i in nonepop_data:
    print('標題    : ',i['title'])
    print('貼文時間 : ',i['createdAt'])
    print('留言人數 : ',i['commentCount'])
    print('按讚人數 : ',i['likeCount'])
    print('=============')
    nonepop_message = nonepop_message + i['commentCount']
    nonepop_like = nonepop_like + i['likeCount']

print('非熱門留言總數',nonepop_message)
print('非熱門按讚總數',nonepop_like)
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

print('popular avg message', pop_message/len(pop_data))
print('popular avg like',pop_like/len(pop_data))
print('======================')
print('none popular avg message', nonepop_message/len(nonepop_data))
print('none popular avg like',nonepop_like/len(nonepop_data))

```
